```python
import PyPDF2
import regex as re
```

```python
file = open('Aposte online na Mega-Sena.pdf', 'rb')

fileReader = PyPDF2.PdfFileReader(file)

print(fileReader.numPages)
```

```python
text = ''
for page in range(fileReader.numPages):
    pageObj = fileReader.getPage(page)
    text = text + pageObj.extractText()
```

```python
bets = re.findall('((?:[0-9] ){11}[0-9])',text)
len(bets)
```

```python
lstbets = []
for item in bets:
    bet = item.replace(' ','')
    bet_number = [int(i) for i in [bet[:2],bet[2:4],bet[4:6],bet[6:8],bet[8:10],bet[10:]]]
    lstbets.append(bet_number)
```

```python
numbers = [22,37,3,41,43,20]
dict_result = {}
index = 0
for bet in lstbets:
    dict_result[index] = (bet, sum(el in bet for el in numbers))
    index = index+1
dict(sorted(dict_result.items(), key=lambda item: item[1][1], reverse=True))
```
